```python
import numpy as np

# create a list of random integers to 100
digits = np.random.randint(-100, 100, size=100)
odd_digits = np.array([])

for i in range(len(digits)):
    if digits[i] % 2 != 0:
        odd_digits = np.append(odd_digits, digits[i])

print(digits)
print(odd_digits)
```

```python
import numpy as np

digits = np.random.randint(0, 100, size=20)

for i in range(len(digits)):
    if digits[i] % 2 != 0:
        digits[i] = -1

print(digits)
```

```python
import numpy as np

matrix_a = np.random.randint(0, 100, size=(5, 1))
print(matrix_a)

matrix_b = np.random.randint(0, 100, size=(5, 1))
print(matrix_b)

concat_matrix = np.concatenate((matrix_a, matrix_b), axis=1)
print(concat_matrix)
```
