2. In các số nguyên tố từ [a,b], a và b là do người dùng nhập vào (b>a)


```python
a = int(input("Nhap a:"))
b = int(input("Nhap b:"))

for num in range(a, b+1):
    is_prime = True
    for i in range(2, int(num*0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
```

3. Kiểm tra xem một số có phải là số nguyên tố hay không?


```python
num = int(input("Nhập một số: "))
is_prime = True
if num < 2:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
if is_prime:
    print(num, "là số nguyên tố.")
else:
    print(num, "không phải là số nguyên tố.")
```

5. Viết 1 hàm để xoá 1 phẩn tử từ dictionary theo key?


```python
# Bài tập 5
def xoa_phan_tu_dictionary(dic, key):
    if key in dic:
        del dic[key]
    return dic

# Kiểm tra bằng cách gọi hàm
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
key_can_xoa = "age"
print(xoa_phan_tu_dictionary(my_dict, key_can_xoa))
```

6. Viết 1 hàm để tạo 1 dictionary từ 2 list, một list chứa keys và 1 list chứa values.


```python
def tao_dictionary(keys, values):
    my_dict = {}
    for i in range(len(keys)):
        my_dict[keys[i]] = values[i]
    return my_dict

# Kiểm tra bằng cách gọi hàm
keys = ["name", "age", "city"]
values = ["Alice", 30, "New York"]
print(tao_dictionary(keys, values))
```

7. Viết 1 hàm để tìm key có giá trị lớn nhất trong 1 dictionary

```python
def tim_key_gia_tri_lon_nhat(dic):
    max_value = max(dic.values())
    max_keys = [key for key, value in dic.items() if value == max_value]
    return max_keys

# Kiểm tra bằng cách gọi hàm
my_dict = {"a": 5, "b": 10, "c": 5, "d": 20}
print(tim_key_gia_tri_lon_nhat(my_dict))
```

8. Viết 1 hàm để đếm số lần xuất hiện của 1 phần tử trong list


```python
# Bài tập 8
def dem_so_lan_xuat_hien(lst, phan_tu):
    dem = 0
    for item in lst:
        if item == phan_tu:
            dem += 1
    return dem

# Kiểm tra bằng cách gọi hàm
my_list = [1, 2, 3, 2, 4, 2, 5]
phan_tu_can_dem = 2
print(dem_so_lan_xuat_hien(my_list, phan_tu_can_dem))
```

Bai 3: Sử dụng cấu trúc từ điển, viết một chương trình nhập vào 1 câu, in số chữ cái và chữ số trong câu đó.

```python
st = input("Nhap vao 1 chuôi:")
d = {"so chu cai": 0,
     "so chu so": 0}

for ch in st:
    if ch.isalpha() == True:
        d["so chu cai"] += 1
    elif ch.isdigit() == True:
        d["so chu so"] += 1
    else: 
        pass

print("So chu cai la:", d["so chu cai"])
print("So chu so la:", d["so chu so"])
```

```python
# Viết chương trình tìm tất cả các số trong đoạn 100 và 300 (tính cả 2 số này) sao cho tất cả
# các chữ số trong đó là số chắn. In ra các số tìm được thành chuỗi cách nhau bởi dấu phẩy
import random
result = []
random_num = [random.randint(100, 301) for i in range(100, 301)]
print(random_num)

for num in random_num:
    if num % 2 == 0:
        result.append(num)

print(result)
```
